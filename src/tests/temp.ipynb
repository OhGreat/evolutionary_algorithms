```python
import sys
import numpy as np
sys.path.append("../")
from EA_components.Population import Population
```

```python
def discrete(parents: Population, offspring: Population):
    # Choose parents
    idxes = range(parents.pop_size)
    parent_choices = np.random.choice(idxes,size=(offspring.pop_size, 2))
    print(parent_choices)
    # Choose probs for each element
    elem_probs = np.random.uniform(size=(offspring.pop_size, offspring.ind_size))
    print(elem_probs)
    new_inds = []
    for i in range(offspring.pop_size):
        p1, p2 = parents.individuals[parent_choices[i]]
        new_ind = [p1[elem] if elem_probs[i,elem] > .5 
                else p2[elem]
                for elem in range(offspring.ind_size)]
        new_inds.append(new_ind)
        #print(new_ind)
    offspring.individuals = np.array(new_inds)
```

```python
p = Population(3,4,"Intermediate")
off = Population(5,4,"Intermediate")
print(p.individuals)
discrete(p,off)
```
